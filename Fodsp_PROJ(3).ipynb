````python
b=sns.PairGrid(Canada, vars=["co2", "trade_co2","other_industry_co2","flaring_co2","gas_co2"],hue="co2")
b.map_diag(sns.histplot)

b.map_offdiag(plt.scatter)
b.add_legend()
plt.savefig('pairgrid.png')
````

**Inference:**

1.  co2 have positive non linear relation with coal_co2, oil_co2, and gas_co2 

2. flaring_co2 is not related to co2,trade_co2,other_industry_co2,gas_co2   
3.   gas_co2 and other_industry_c02 have negative non linear relation
    





````python
b=sns.PairGrid(Canada, vars=["co2", "co2_per_gdp","consumption_co2_per_gdp","cumulative_co2","population"],hue='co2')
b.map(sns.scatterplot)
plt.savefig('pairgridnohist.png')
````

**Inference:**

1.   cummulative_co2 and population of a country are linearly related to each other
2.  there is negative linear relation between cumulative_co2 and co2_per_gdp
